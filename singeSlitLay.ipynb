```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ----- Parameters -----
L = 2.0  # change as needed

# Define ranges for m and n
m_values = range(0, 10)   # example: m = 1..3
n_values = range(0, 10)   # example: n = 1..3

# x range
x = np.linspace(0, 10, 1000)
```

```python
# ----- Define functions -----
def a(m, n, L):
    return (m**2 - n**2) / (4 * L)

def c(m, n):
    return 0.5 * (m - n)
```

```python
for m in m_values:
    for n in n_values:
        if m == n:
            continue

        a_val = a(m, n, L)
        c_val = c(m, n)

        if a_val == 0:
            continue

        radicand = c_val**2 - a_val**2
        if radicand <= 0:
            continue  # skip non-real hyperbolas

        y = np.sqrt(radicand) * np.sqrt(1 + x**2 / a_val**2)

        plt.plot(x, y, label=f"m={m}, n={n}")

# ----- Formatting -----
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Hyperbolas for different (m,n), m ≠ n")
plt.grid(True)
plt.legend()
plt.show()
```
