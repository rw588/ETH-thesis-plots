```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-muted')
amp_color   = "#4C72B0"   # muted blue
trans_color = "#DD8452"   # muted orange
cont_color  = "#55A868"   # muted green
```

```python
def amplitude(f):
    return f**2 * np.abs(np.sinc(f) * np.sinc(2*f))

#calculate contrast by calculating the transmission
def transmission(f):
    return f**3

#calculate contrast by max-min/max+min
def contrast(f):
    A = amplitude(f)
    T = transmission(f)
    return np.where(A > T, 1.0, A / T)

```

```python
x = np.linspace(0, 1, 1000)
plt.plot(x, amplitude(x))
plt.xlabel('Open fraction')
plt.ylabel('Amplitude')
plt.title('Amplitude of a deflectometer')
plt.grid()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# Amplitude (left axis)
ax1.plot(x, amplitude(x), color=amp_color, linewidth=2)
ax1.set_ylabel("Amplitude", color=amp_color)
ax1.tick_params(axis='y', labelcolor=amp_color)

# Transmission (first right axis)
ax2 = ax1.twinx()
ax2.plot(x, transmission(x), color=trans_color, linewidth=2)
ax2.set_ylabel("Transmission", color=trans_color)
ax2.tick_params(axis='y', labelcolor=trans_color)

# Contrast (second right axis)
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(x, contrast(x), color=cont_color, linewidth=2)
ax3.set_ylabel("Contrast", color=cont_color)
ax3.tick_params(axis='y', labelcolor=cont_color)

plt.title("Amplitude, Transmission, and Contrast of a Deflectometer")
plt.grid()
plt.show()
```
